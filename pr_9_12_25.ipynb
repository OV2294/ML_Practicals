```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
X = [[1],[2],[3],[4]]
Y = [2,4,6,8]
model.fit(X,Y)
prediction = model.predict([[10]])
print("Prediction for input 5 is :", prediction)
```

User Input Prediction

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
X = [[1],[2],[3],[4]]
Y = [2,4,6,8]
model.fit(X,Y)
#take user input
user_value = float(input("Enter a numnber to predict: "))
#predict
prediction = model.predict([[user_value]])
print("Prediction is :", prediction[0])
```

Visualize

```python
import matplotlib.pyplot as plt
#Scatter plot(data points)
plt.scatter(X,Y, color='blue', label='Data Points')
#Regression line
#plt.plot(X, model.predict(X), color='orange', label='Regression Line')
#Add legend 
plt.legend(loc='center left',bbox_to_anchor=(1,0.5))
plt.title("Linear Regression Line")
plt.xlabel("x")
plt.ylabel("y")
plt.tight_layout() # keeps layout neat when utside
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

model1 = LinearRegression()

weights = np.array([1,2,3,4,5]).reshape(-1,1) #features
prices = np.array([30,60,90,120,150]) #labels
# print(weights.size)
# print(prices.size)
model1.fit(weights,prices)
#Scatter plot(data points)
plt.scatter(weights,prices, color='blue', label='Data Points')
#Regression line
plt.plot(weights, model1.predict(weights), color='orange', label='Regression Line')
#legend 
plt.legend(loc='center left',bbox_to_anchor=(1,0.5))
#title for graph
plt.title("Linear Regression Line")
# x and y axis labels
plt.xlabel("weights")
plt.ylabel("prices")
plt.tight_layout() # keeps layout neat when utside
plt.show()
```
